```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
data = pd.read_csv("../../Dataset/FINAL_LinkedCleanSalesWeatherWithEncoding.csv")
```

```python
data.head()
```

```python
drop_data = data.drop("date", axis = 1)
cor_matrix = drop_data.corr()
cor_matrix
```

```python
%matplotlib inline

mp.rcParams['figure.figsize'] = 30,20
mp.rcParams['xtick.labelsize'] = 20
mp.rcParams['ytick.labelsize'] = 20

corr_data = sb.heatmap(cor_matrix, cmap = 'YlGnBu', linewidths = 0.5)
corr_data
```
